# Feature Engineering

# Data Cleaning

```python
#dropping patient_id because it is not so used for model building
df.drop(['Patient_ID'],axis=1,inplace=True)
```

```python
#fill na values
df['Formulation'].fillna(df['Formulation'].mode()[0],inplace=True)
df['DrugName'].fillna(df['DrugName'].mode()[0],inplace=True)
df['SubCat'].fillna(df['SubCat'].mode()[0],inplace=True)
df['SubCat1'].fillna(df['SubCat1'].mode()[0],inplace=True)
```

```python
#all missing values are filled 
df.isna().sum()
```

```python
#removing duplicate values
df.drop_duplicates(inplace=True)
```

```python
# all duplicated values are removed
df.duplicated().sum()
```

```python
#From above boxplot we got to know in our data outliers are present
#so now removing the outlier
# removal of outliers are important because, when outliers are present it will affect the data
```

```python
df_num
#it is the dataframe in which all our numeric data are present. So this numerical data containing outliers which is affecting our model.
```

```python
#code for removal of outliers
num=[f for f in df.columns if df[f].dtypes!='O']
for i in num:
    q3=np.percentile(df[i],75)
    q1=np.percentile(df[i],25)
    iqr=q3-q1
    ub=q3+1.5*iqr
    lb=q1-1.5*iqr
    
    df[i]=np.where(df[i]>ub,ub,df[i])
    df[i]=np.where(df[i]<lb,lb,df[i])
```

```python
#after removal of outliers
#boxplot
#boxplot with no outliers
df.plot(kind='box',figsize=(9,7),subplots=True,layout=(3,2),sharex=False,sharey=False)
plt.show()
```

# Label Encoding

```python
# so for model building we have to convert in numerics
# It is done using LabelEncoding
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# label encoding using for loop
l=[i for i in df.columns if df[i].dtypes=='O']
for i in l:
    df[i]=le.fit_transform(df[i])
```
